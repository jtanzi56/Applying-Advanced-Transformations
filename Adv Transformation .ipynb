```python
df_hero['Measurements'] = df_hero['Measurements'].str.replace("'",'"')
df_hero['Measurements'] = df_hero['Measurements'].apply(json.loads)
df_hero['Measurements'].head()
```

```python
height_weight = df_hero['Measurements'].apply(pd.Series)
height_weight
```

```python
df_hero = pd.concat((df_hero, height_weight), axis = 1)
df_hero.head()
```

```python
df_hero = df_hero.drop(columns=['Measurements'])
```

```python
df_hero.info()
```

```python
df_hero['Height'].unique()
```

```python
to_replace = [' cm', '']

for char in to_replace:
    df_hero['Height'] = df_hero['Height'].str.replace(char,'',regex=False)
```

```python
df_hero['Height'].head()
```

```python
df_hero['Height'] = pd.to_numeric(df_hero['Height'])
```

```python
df_hero.rename(columns={'Height': 'Height (cm)'}, inplace=True)
```

```python
df_hero.head()
```

```python
to_replace = [' kg', '']

for char in to_replace:
    df_hero['Weight'] = df_hero['Weight'].str.replace(char,'',regex=False)
df_hero['Weight'] = pd.to_numeric(df_hero['Weight'])
df_hero.rename(columns={'Weight': 'Weight (kg)'}, inplace=True)
df_hero.head()
```

```python
exploded = df_pow.explode('Powers_split')
exploded[['hero_names','Powers','Powers_split']].head(5)
```

```python
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df_pow[col] = df_pow['Powers'].str.contains(col)
df_pow.head()
```

```python
df_pow.head()
```

```python
hero_pow_df = pd.merge(df_hero, df_pow, left_on = 'Hero', right_on = 'hero_names')
```
